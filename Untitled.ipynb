```python
import skimage as ski
import numpy as np


```

```python



image = ski.img_as_float(ski.io.imread("535.jpg"))
image_resized = ski.transform.resize(image, (int(image.shape[0]/10), int(image.shape[1]/10)))
ski.io.imshow(image_resized)
rotate = ski.transform.rotate(image_resized, 45, mode='constant', cval=0.1020544873874387438)
c,r,k=rotate.shape
for i in range(c):
    for j in range(r):
        if np.allclose(rotate[i][j],[0.1020544873874387438]*3):
            rotate[i][j]=[0,1,0]
ski.io.imshow(rotate)
```
